# Assign each to a to a variable called cars1 and cars2
-data 
    - cars1.csv and cars2.csv

```python
import pandas as pd

cars1 = df = pd.read_csv("cars1.csv")
cars1
```

```python
cars2 = df = pd.read_csv("cars2.csv")
cars2
```

#  fix cars1 having some unnamed blanks ie: take only the column except unnamed

```python
cars1 = cars1.drop(columns=cars1.loc[:,'Unnamed: 9':'Unnamed: 13'])
cars1
```

 # What is the number of observations in each dataset?

```python
print(cars1.shape,'\n',cars2.shape)

print(cars1.count(axis=0),'\n')

print(cars2.count(axis=0))
```

# Join cars1 and cars2 into a single DataFrame called cars

```python
cars = pd.concat([cars1,cars2])
cars
```

#  Create a random number Series from 15,000 to 73,000. as owners

```python
import numpy as np
np.random.seed()
owners = np.random.randint(15000,73000,cars.shape[0])
owners
```

# Add the column owners to cars


```python
cars['owners'] = owners
cars
```

 # Fictitious Names

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
data1 = pd.DataFrame(raw_data_1, columns = ['subject_id', 'first_name', 'last_name'])
data2 = pd.DataFrame(raw_data_2, columns = ['subject_id', 'first_name', 'last_name'])
data3 = pd.DataFrame(raw_data_3, columns = ['subject_id','test_id'])

data3
```

#  Join the two dataframes along rows and assign all_data

```python
all_data = pd.concat([data1,data2],axis=0)
all_data
```

# Join the two dataframes along columns and assing to all_data_col

```python
all_data_col = pd.concat([data1,data2],axis=1)
all_data_col
```

# Merge all_data and data3 along the subject_id value

```python

merger = all_data.merge(data3,on='subject_id')
print(all_data, '\n')
print(data3,'\n')
print(merger)
```

# Merge only the data that has the same 'subject_id' on both data1 and data2

```python
merger2 = data1.merge(data2, on='subject_id',how='inner')
print(data1, '\n')
print(data2,'\n')
print(merger2)
```

# Merge all values in data1 and data2, with matching records from both sides where available.

```python
merger3 = data1.merge(data2, on='subject_id',how='outer')
print(data1, '\n')
print(data2,'\n')
print(merger3)
```
